```python
# Scipy:
# We have the min and max temperatures in a city In India for each months of the year.
# We would like to find a function to describe this and show it graphically, the dataset
# given below.
# Task:
# 1. fitting it to the periodic function
# 2. plot the fit
# Data
# Max = 39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25
# Min = 21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18
# Expected Output:
```

```python
import numpy as np

max_temp = np.array([39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25])
min_temp = np.array([21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18])

import matplotlib.pyplot as plt
months = np.arange(12)
plt.plot(months, max_temp, 'ro')
plt.plot(months, min_temp, 'bo')
plt.xlabel('Month')
plt.ylabel('Min and max temperature')
```

```python
#fitting to periodic function
from scipy import optimize
def yearly_temps(times, avg, ampl, time_offset):
    return (avg
            + ampl * np.cos((times + time_offset) * 2 * np.pi / times.max()))

res_max, cov_max = optimize.curve_fit(yearly_temps, months,
                                      max_temp, [20, 10, 0])
res_min, cov_min = optimize.curve_fit(yearly_temps, months,
                                      min_temp, [-40, 20, 0])
#plotting the fit
days = np.linspace(0, 12, num=365)

plt.figure()
plt.plot(months, max_temp, 'ro')
plt.plot(days, yearly_temps(days, *res_max), 'r-')
plt.plot(months, min_temp, 'bo')
plt.plot(days, yearly_temps(days, *res_min), 'b-')
plt.xlabel('Month')
plt.ylabel('Temperature ($^\circ$C)')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
url='https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv'
titanic = pd.read_csv(url)
(titanic.head())
```

```python
# 1. Create a pie chart presenting the male/female proportion
titanic.groupby('sex').size().plot(kind='pie',figsize = (6,6),autopct = '%.2f',fontsize=10,shadow=True)
```

```python
mini_df =(titanic[['sex','age','fare']])
# removing the last row with null values
mini_df= mini_df[:-1]
mini_df

mini_df['sex'].replace('female',0,inplace=True)
mini_df['sex'].replace('male',1,inplace=True)
mini_df

fig,ax =plt.subplots(figsize=(15,6))
mini_df.plot(kind ='scatter',x = 'age',
                     y = 'fare',c='sex',cmap='RdYlBu',
                     xticks=[0,5,10,15,20,25,30,35,40,45,50,55,60,65,70,75,80],ax=ax,legend='sex')
```
